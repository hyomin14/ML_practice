```python
#DataFrame 데이터세트 생성과 수정
titanic_df['Family_No'] = titanic_df['SibSp']+ titanic_df['Parch'] + 1
titanic_df.head(3)
```

```python
data = {'Name':['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
       'Year':[2011,2016,2015,2015],
       'Gender':['Male', 'Female', 'Male', 'Male']}
data_df = pd.DataFrame(data, index=['one', 'two', 'three', 'four'])
print(data_df)

data_df_reset = data_df.reset_index()
data_df_reset.index += 1
print(data_df_reset)
```

```python
#ix[행(index 값), 열(칼럼 명, 칼럼 위치 값)] --> 쓰지 않을 것을 권장

#iloc[] --> 위치 기반 인덱싱(integer), 불린 인덱싱 X
print(data_df.iloc[0,0])
print(data_df.iloc[0:1,1])    #1개 행 출력
print(data_df.iloc[0,1])

#loc[] --> 명칭 기반 인덱싱(index 이름)
print(data_df.loc['one', 'Name'])
print(data_df_reset.loc[1, 'Name'])
print(data_df_reset.loc[1:2, 'Name'])    #2개 행 출력
```

```python
#aggregation 함수 --> max(), min(), sum(),...
print(titanic_df.count())
print(titanic_df[['Age', 'Fare']].mean())
```

```python
#groupby() --> by:대상 칼럼으로 groupby, aggregation 함수 호출하면 groupby() 대상 칼럼 제외한 모든 칼럼에 대해 적용
titanic_groupby = titanic_df.groupby(by='Pclass').count()
#titanic_groupby = titanic_df.groupby(by='Pclass')[['PassengerId', 'Survived']].count()
titanic_groupby
```

```python
#적용하려는 여러개의 aggregation 함수를 agg() 내에 인자로 입력
titanic_df.groupby('Pclass')['Age'].agg([max, min])
```

```python
#여러개의 칼럼에 서로 다른 aggregation 함수를 groupby에서 호출하려면 agg()이용
agg_format = {'Age':'max', 'SibSp':'sum','Fare':'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
```

```python
#결손 데이터 : 칼럼에 값이 없는, 즉 NULL인 경우, 넘파이의 NaN으로 표시 --> 평균, 총합 등의 연산 시 제외됨
#결손 데이터 처리 --> isna(), fillna()
titanic_df.isna().head()
```

```python
#결손데이터 개수
titanic_df.isna().sum()
```

```python
#결손 데이터 대체
```

```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head()
#반환 데이터를 다시 저장해주거나 inplace=True로 설정
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
titanic_df.isna().sum()
```

```python
#lambda : 함수의 선언과 함수 내의 처리를 한 줄의 식으로 변환
lambda_square = lambda x : x ** 2
print('3의 제곱은 :', lambda_square(3))

#여러개의 값을 입력 인자로 --> map()
a = [1,2,3]
squares = map(lambda x : x ** 2, a)
list(squares)
```

```python
#apply() --> 특정 행 또는 데이터프레임 전체에 함수 적용
#apply lambda 식으로 데이터 가공
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']]
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']]
#else if는 지원하지 않기 때문에 else절에 if-else절을 내포한 괄호를 적용
```

```python
#나이에 따라 세분화된 분류 수행하는 함수 생성
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else: cat = 'Elderly'
    
    return cat

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']]
```

데이터를 전처리하고 적절한 피처를 가공/추출하는 부분이 많은 비중을 차지하기 때문에 넘파이, 판다스 등 다양한 패키지에 대한 이해 필요
